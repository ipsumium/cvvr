```python
#Stergere item
def itemdeletep(database,index):
    filename=database.name
    database.close
    tbl = dbfdd.Table(filename)
    tbl.open(mode=dbfdd.READ_WRITE)
    rec = tbl[index]
    dbfdd.delete(rec)
    tbl.pack()
    tbl.close()
    del rec
    return dbfb.Dbf(filename)
```
